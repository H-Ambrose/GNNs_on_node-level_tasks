## comparison: Cora

```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
data_gcn = np.loadtxt("LossAndAccuracy/GCN_Cora.txt")
data_gat = np.loadtxt("LossAndAccuracy/GAT_Cora.txt")
data_sage = np.loadtxt("LossAndAccuracy/GraphSAGE_Cora.txt")

x = data_gcn[:,0] # epoch

y_gcn = data_gcn[:,2] # accuracy
y_gat = data_gat[:,2] # accuracy
y_sage = data_sage[:,2] # accuracy
```

```python
fig = plt.figure(figsize = (7,5))       #figsize是图片的大小`

pl.plot(x, y_sage, 'c-', label = u'GraphSAGE')
pl.legend()

pl.plot(x, y_gcn,'g-',label = u'GCN')
pl.legend()

pl.plot(x, y_gat,'y-',label = u'GAT')
pl.legend()

pl.xlabel(u'epochs')
pl.ylabel(u'loss')
plt.title('Compare accuracy for different models on cora')

plt.savefig("graphs/com_cora.png")
```
